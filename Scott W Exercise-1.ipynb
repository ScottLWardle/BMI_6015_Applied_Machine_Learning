## PART A - NumPy

* **(10 points)** Create a random dataset with 1000 rows and 6 columns by generating the columns one at a time and then concatenating them to create the dataset.

```python
# Start with a repeatable random seed
rgen = np.random.RandomState(73)
# create 6 2D column vectors
array1 = rgen.randint(1, 7, size=(1000, 1))
array2 = rgen.normal(loc=10, scale=5, size=(1000,1))
array3 = rgen.randn(1000,1)
array4 = rgen.randint(6, 20, size=(1000, 1))
array5 = rgen.normal(loc=0, scale=10, size=(1000,1))
array6 = rgen.normal(loc=3, scale=1, size=(1000,1))
# combine them all inta a dataset
rand_dataset = np.concatenate([array1, array2, array3, array4, array5, array6], axis=1)
print(rand_dataset.shape)
```

* **(8 points)** Shuffle the dataset.

```python
# Create a range of indices the same length of the rows in dataset
indexes = np.arange(len(rand_dataset))
np.random.shuffle(indexes)
indexes
```

```python
# Use some fancy indexing to shuffle the random dataset
new_rand_ds = rand_dataset[indexes]
new_rand_ds
```

* **(10 points)** Split it using a 60/25/15 split into three sets: training, validation and test.

```python
## Split into three datasets
split_ndx1 = int(.6 * len(new_rand_ds))
split_ndx2 = int(.85 * len(new_rand_ds))
```

Split with the 60% of the records:

```python
training_ds = new_rand_ds[0:split_ndx1, :]
training_ds
```

Split with the 25% of the records:

```python
validation_ds = new_rand_ds[split_ndx1:split_ndx2, :]
validation_ds
```

Split with 15% of the records:

```python
test_ds = new_rand_ds[split_ndx2:, :]
test_ds
```

* **(7 points)** Standardize every column in the training set by subtracting every column by its mean and dividing the result by its standard deviation.

```python
# Create an array of all the means for each column from the training array
mean_cols_train=training_ds.mean(axis=0)
mean_cols_train
```

```python
# Create an array of all the standard deviations of each column from training aray
std_cols_train=training_ds.std(axis=0)
std_cols_train
```

```python
train_ds_stdz=(training_ds-mean_cols_train)/std_cols_train
train_ds_stdz
```

* **(5 points)** Do the same with validation and test sets but using the mean and standard deviation from the training set.

```python
# Standardize the validation set with train means and std
valid_ds_stdz=(validation_ds-mean_cols_train)/std_cols_train
valid_ds_stdz
```

```python
# Standardize the test set with the train means and std
test_ds_stdz=(test_ds-mean_cols_train)/std_cols_train
test_ds_stdz
```

## PART C - Matplotlib

* **(27 points)** Plot the following functions using $2\ by\ 3$ subplots. Each plot must have its own title, and its own x and y labels. Make also sure that each plot enough of the function to show its behavior.
    * $f(x) = 3 x^2 + 10$
    * $f(x) = \frac{1}{1 + e^{-x}}$
    * $f(z) = max(\alpha z, z)$ where $\alpha = 0.01$ where $max(...)$ returns the maximum of its arguments.
    * $f(x) = sin(x^2)$
    * $f(x) = cos(x) + ln(x)$
    * $f(x) = tanh(x)$

```python
# Define figure for a 2x3 plotting space
fig = plt.figure(figsize=(24,9))

ax = fig.add_subplot(2, 3, 1) # First subplot
x = np.arange(1, 10 ,0.1)
ax.plot(x, 3*x**2 +10, 'b-')  
ax.set_ylabel("$3x^2 + 10$")
ax.set_xlabel('x')
ax.set_title("Polynomial function")
ax.grid()

ax = fig.add_subplot(2, 3, 2) # Second subplot
ax.plot(x, 1 / (1 + np.exp(-x)), 'r-')
ax.set_ylabel("$1/1 + exp(-x)$")
ax.set_xlabel('x')
ax.set_title("Exponential function")
ax.grid()

alpha = 0.01
z = np.arange(-5, 6, 1)
ax = fig.add_subplot(2, 3, 3) # Third subplot
ax.plot(z, np.maximum(alpha*z,z), 'g-')
ax.set_ylabel(r"$\max(\alpha z, z)$")
ax.set_xlabel('z')
ax.set_title("Piecewise linear function")
ax.grid()

ax = fig.add_subplot(2, 3, 4) # Fourth subplot
ax.plot(x, np.sin(x**2), 'm-')
ax.set_ylabel("$sin(x^2)$")
ax.set_xlabel('x')
ax.set_title("Sinsiodal function")
ax.grid()

ax = fig.add_subplot(2, 3, 5) # Fifth subplot
ax.plot(x, np.cos(x) + np.log(x), 'o-')
ax.set_ylabel("$cos(x) + log(x)$")
ax.set_xlabel('x')
ax.set_title("Transendental function")
ax.grid()

ax = fig.add_subplot(2, 3, 6) # Sixth subplot
ax.plot(x, np.tanh(x), 'y-')
ax.set_ylabel("$tanh(x)$")
ax.set_xlabel('x')
ax.set_title("Hyperbolic tangent function")
ax.grid()
plt.show()
```
